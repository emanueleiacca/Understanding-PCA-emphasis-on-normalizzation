```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler                      # For normalizing Data
import seaborn as sns
from sklearn.decomposition import PCA
```

```python

column_names = [
    "Alcohol", "Malic acid", "Ash", "Alcalinity of ash", "Magnesium",
    "Total phenols", "Flavanoids", "Nonflavanoid phenols", "Proanthocyanins",
    "Color intensity", "Hue", "OD280/OD315 of diluted wines", "Proline"
]

wine = pd.read_csv('/content/wine.data', names=column_names)
```

```python
wine_scaled = StandardScaler().fit_transform(wine)
pca = PCA()
X_new = pca.fit_transform(wine_scaled)
pca.get_covariance()
```

```python
explained_variance=pca.explained_variance_ratio_
explained_variance
```

```python
cumulative_variance = explained_variance.cumsum()

plt.bar(range(1, len(explained_variance) + 1), explained_variance, alpha=0.7, align='center', label='Varianza spiegata')
plt.step(range(1, len(cumulative_variance) + 1), cumulative_variance, where='mid', label='Varianza cumulativa spiegata')
plt.xlabel('Numero di Componenti Principali')
plt.ylabel('Varianza Spiegata')
plt.legend(loc='best')
plt.title('Varianza Spiegata e Cumulativa Spiegata')
plt.show()
```

```python
#elbow method
n_components = len(explained_variance)
components = np.arange(1, n_components + 1)
cumulative_variance = np.cumsum(explained_variance)

plt.figure(figsize=(8, 6))
plt.plot(components, explained_variance, marker='o', linestyle='-', label='Explained Variance')
plt.plot(components, cumulative_variance, marker='o', linestyle='-', label='Cumulative Explained Variance')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance')
plt.title('Explained Variance vs. Number of Components')
plt.grid(True)
plt.legend()

threshold = 0.8

elbow_index = np.argmax(cumulative_variance >= threshold)

optimal_num_components = elbow_index + 1 #index in py start from 0
print(f"Optimal Number of Components: {optimal_num_components}")

plt.vlines(optimal_num_components, plt.ylim()[0], plt.ylim()[1], linestyle='--', label='Elbow Point', color='red')
plt.legend()

plt.show()


```

```python
def standardize(x):
    return (x - np.mean(x)) / np.std(x)

def rescale(x):
    return (x - np.min(x)) / (np.max(x) - np.min(x))

def index(x, base=100):
    return (x / np.mean(x)) * base
#I choosed to define the function because I had it already done from an other study, you can import it from sklearn
```

```python
wine_sd = wine.apply(lambda x: standardize(x))
wine_rs = wine.apply(lambda x: rescale(x))
wine_in = wine.apply(lambda x: index(x))
```

```python
pca = PCA()
X_new = pca.fit_transform(wine_sd)
pca.get_covariance()
```

```python
explained_variance=pca.explained_variance_ratio_
explained_variance
```

```python
pca = PCA()
X_new = pca.fit_transform(wine_rs)
pca.get_covariance()
```

```python
explained_variance=pca.explained_variance_ratio_
explained_variance
```

```python
pca = PCA()
X_new = pca.fit_transform(wine_in)
pca.get_covariance()
```

```python
explained_variance=pca.explained_variance_ratio_
explained_variance
```

We wanted to explore the difference between different method of Standardizzation, we can see that each method give us a different explained variance. The best method depends by the purpose of our studies, it's just important to know the difference
